# Creating boxplot for Age

```python
plt.figure(figsize =(10, 7))
plt.boxplot(df['Age'], showmeans=True, vert=False, manage_ticks=False)
plt.ylabel('Age')
plt.title('Box Plot for Age')
plt.show()
```

# Treating missing values present in "Height" before creating boxplot

```python
df['Height'] = df['Height'].fillna(df['Height'].median())
```

# Creating box plot for Height

```python
plt.figure(figsize =(10, 7))
plt.boxplot(df['Height'], showmeans=True, vert=False, manage_ticks=False)
plt.ylabel('Height')
plt.title('Box Plot for Height')
plt.show()
```

# Treating missing values present in "Weight" before creating boxplot

```python
df['Weight'] = df['Weight'].fillna(df['Weight'].median())
```

# Creating box plot for Weight

```python
plt.figure(figsize =(10, 7))
plt.boxplot(df['Weight'], showmeans=True, vert=False, manage_ticks=False)
plt.ylabel('Weight')
plt.title('Box Plot for Weight')
plt.show()
```

# Creating a scatter plot to show relationship b/w age and weight

```python
plt.scatter(df['Age'], df['Weight'])
plt.xlabel('Age')
plt.ylabel('Weight')
plt.title('Scatter Plot')
plt.show()
```

# Inferences drawn from the above visualisations

1. Age is having lot of outliers and is heavily right skewed distribution

2. Height has outliers on both side but it is balanced and it shows normal distribution.

3. Weight has more outliers on right side thus making it right skewed.

4. Not much relation could be seen between Age and Weight but there seem to be slight positive correlation.

# Sentiment analysis

```python
#Creating a new column for sentiment analysis and figuring out the proportion of each sentiment
```

```python
import pandas as pd

import numpy as np

import seaborn as sns

import matplotlib.pyplot as plt


conditions = [

    (df['Medal'] >1),

    (df['Medal'] == 0),

    (df['Medal']==1)]

choices = ['Positive', 'Negative','Neutral'] #Positive is encoded to 1 and negative to 0

df['Sentiment'] = np.select(conditions,choices, default='Negative')


df['Sentiment'].value_counts(normalize=True)
```

Creating pie-chart to showcase the sentiment distribution

```python
plt.figure(figsize=(9,9))

plt.pie(df["Sentiment"].value_counts(),labels=df["Sentiment"].value_counts().index,autopct='%1.1f%%')

plt.show()
```

Doing sentiment analyis for the below condition-
1. athletes who won 1 medal have neutral sentiment
2. athletes who won 0 medal have negaive sentiment
3. athletes who won more than 2 medals have positive sentiment

```python
conditions = [

    (df['Medal'] >1),

    (df['Medal'] == 0),

    (df['Medal']==1)]

choices = ['Positive', 'Negative','Neutral'] #Positive is encoded to 1 and negative to 0

df['Sentiment'] = np.select(conditions,choices, default='Negative')
plt.figure(figsize=(9,9))

plt.pie(df["Sentiment"].value_counts(),labels=df["Sentiment"].value_counts().index,autopct='%1.1f%%', startangle = 90, shadow=True)

plt.legend(title ='Sentiment')

plt.show()
```

Creating two separate datasets taking Judo and Taekwondo as the sport and including the top 3 participating nations for each of them

```python
import pandas as pd
df = pd.read_csv("/content/Olympics.csv")
df_judo = df[df['Sport']=='Judo']

df_judo['Team'].value_counts()

df_judo_top3 = df_judo.loc[df_judo['Team'].isin(['Germany','France','Japan'])]

df_judo_top3.head()
```

```python
df_tae = df[df['Sport']=='Taekwondo']

df_tae['Team'].value_counts()

df_tae_top3 = df_tae.loc[df_tae['Team'].isin(['China','Australia','Korea, Rep.'])]

df_tae_top3.head()
```

Analyzing the height distribution for the above datasets based on gender with the help of violin plot  

```python
sns.set_style("whitegrid", {'axes.grid' : False})  # Optional: Set the plot style

plt.figure(figsize=(10, 6))  # Optional: Set the figure size


sns.violinplot(x="Sex", y="Height", data=df_judo_top3, hue='Team', dodge=True,saturation =0.5, color='Yellow')


# Optional: Add titles and labels

plt.title("Violin Plot for Judo as sport ")

plt.xlabel("Gender")

plt.ylabel("Height")

plt.show()
```

```python
sns.set_style("whitegrid", {'axes.grid' : False})  # Optional: Set the plot style

plt.figure(figsize=(10, 6))


sns.violinplot(x="Sex", y="Height", data=df_tae_top3, hue='Team', dodge=True,saturation =0.5, color='Yellow')


# Optional: Add titles and labels

plt.title("Violin Plot for Taekwando as sport")

plt.xlabel("Gender")

plt.ylabel("Height")

plt.show()
```
